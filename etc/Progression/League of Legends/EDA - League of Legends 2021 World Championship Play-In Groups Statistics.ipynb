```python
import numpy as np
import pandas as pd

import seaborn as sns
```

```python
df = pd.read_csv('League of Legends 2021 World Championship Play-In Groups Statistics.csv')
df
```

```python
def desc_data(df):
    df_describe = []
    for i in df.columns: 
          df_describe.append([
          i,
          df[i].dtypes,
          df[i].isna().sum(),
          round((((df[i].isna().sum())/(len(df)))*100),2),
          df[i].nunique(),
          df[i].unique() 
      ])
    return pd.DataFrame(df_describe, columns=['column','type','nan','nan%','nunique','unique'])
display(desc_data(df))
print('Total duplicated data:',df.duplicated().sum())
```
