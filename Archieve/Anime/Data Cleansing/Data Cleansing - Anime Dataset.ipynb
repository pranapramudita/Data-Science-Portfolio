# Data Cleansing - Anime Dataset

```python
import pandas as pd
import numpy as np
```

## User Rating Dataset

```python
df = pd.read_csv('rating_complete.csv')
df
```

```python
df[df['user_id'] < 10]['user_id'].value_counts()
```

## User Rating (user_id=[1,2,4,7])

```python
df2 = df.set_index('user_id').loc[[4,1,7,2]]
```

```python
df2 = df2.reset_index().sort_values(by='user_id').reset_index(drop=True)
```

```python
df2
```

```python
df2.user_id.value_counts()
```

```python
df2
```

```python
df2.to_csv('user_rating.csv',index=False)
```

## Anime Dataset

```python
df3 = pd.read_csv('anime.csv',usecols=['MAL_ID','Name','Genres'])
df3 = df3.rename(columns={"MAL_ID":"anime_id",'Name':'anime_name','Genres':'genres'})
df3
```

```python
df3.to_csv('anime_data.csv',index=False)
```
