```python
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
df['Male'] = df.Sex.apply(lambda x: 1 if x == 'M' else 0)
```

```python
df.ChestPainType.unique()
```

```python
df['ChestPain'] = df.ChestPainType.apply(lambda x: 0 if x=='ATA' else(1 if x=='NAP' else(2 if x=='ASY' else 4)))
```

```python
df
```

```python
print(df.Sex.value_counts())
print(df.Male.value_counts())

print(df.ChestPainType.value_counts())
print(df.ChestPain.value_counts())
```
